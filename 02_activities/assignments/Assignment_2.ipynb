**Good Data Visualization**

![Good](participation/images/Good.png)
Available at: https://www.nytimes.com/interactive/2015/07/08/us/census-race-map.html?_r=1

The New York Times interactive visualization Mapping Segregation (2015) is a good example of effective data visualization because it presents demographic data within geographic context, allows for perceptual interpretation, and uses visualization to convey a substantive narrative.

The visualization is effective in placing data in spatial context, through mapping segregation by geographic position to reveal the spatial distribution of racial groups across key cities in the United States. The dot-density representation allows for immediate identification of clustering and separation, allowing viewers to see patterns of residential segregation rather than observing aggregate counts (ie versus a tabular format) .

The visualization also performs well across the three core quality dimensions emphasized in the course: aesthetic, substantive, and perceptual. From an aesthetic standpoint, the two-dimensional layout and restricted colour palette support visual consistency and minimizes distraction. Substantively, the visualization maintains underlying census data with self-reported race categorizations. Perceptually, the visualization conveys a clear and interpretable message: residential segregation is widespread and systematically structured rather than randomly distributed. The visualization is also interactive with zoom functionality allowing examination of patterns at multiple geographic scales without being too complicated.

It is also clear that this data visualization is a tool for advocacy. I think the visualization focuses attention on a social phenomenon of public relevance. This aligns with Data Feminism Chapter 3’s argument that visualizations are inherently rhetorical and shaped by design choices. Importantly, while the visualization is not neutral, it remains transparent in its goal and substantively accurate.

While the visualization is strong overall, a potential improvement would be the inclusion of contextual annotations. As discussed in Data Feminism, critiques of “mythical, objective viewpoints” emphasize that data are never produced from nowhere, but reflect historical and structural conditions. Given the topic, brief annotations referencing racially restrictive zoning bylaws and municipal planning practices that institutionalized segregation (see  https://urbanland.uli.org/a-brief-history-of-racial-zoning-neighborhood-associations-and-municipal-zoning) could provide context for present-day spatial patterns without compromising substantive accuracy.



**Bad Data Visualization**

![Bad](participation/images/Bad.png)

Available at: https://www.nytimes.com/2022/01/06/opinion/omicron-covid-us.html

The spiral visualization used in the New York Times opinion piece on the Omicron wave of COVID-19 (2022) is a poor example of data visualization. The visual is not aligned well with its analytical purpose, requires increased cognitive load for interpretation, and prevents accurate temporal comparisons.

The visualization relies on unfamiliar perceptual representations for quantitative comparisons.  As discussed in the lecture, position along a common linear scale is the most accurate visual encoding for numerical values. However, in this visualization the case counts are using a radial distance along a spiral, which makes interpretation very challenging. This further undermines substantive clarity and complicates meaningful comparison across pandemic waves.

Because of this choice it increases cognitive load, which is identifed as one of the most controllable elements of visualization design. The spiral chart is an unfamiliar and rare chart type that requires viewers to trace a curved path, retain prior values in memory, and compare magnitudes across varying radii. Rather than supporting intuitive interpretation, the design places increased burden on working memory, violating the guidance that effective visualizations should rely on familiar formats and minimize interpretive effort, particularly for general audiences.

Lastly the visualization is poorly aligned with its intended purpose, audience, and choice of visualization. The stated purpose is explanatory—to communicate how the Omicron surge compares with prior COVID-19 waves for a broad public audience. However, the spiral format is complex and requires a high cognitive load and sustained interpretive effort. This mismatch reduces accessibility and weakens the visualization’s effectiveness as a public communication tool.

I think a simple but effective improvement to the visualization would be to use a familiar two-dimensional time-series visualization to reduce cognitive load. Replacing the spiral with a standard line chart, with time on the x-axis and case counts on the y-axis, would align with cognitive efficiency, allow for direct comparison between waves, and allow for clearer interpretation.
