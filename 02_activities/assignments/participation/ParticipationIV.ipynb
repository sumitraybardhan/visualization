```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests


np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))
ax1.scatter(x1,y1)
ax1.annotate ('This is important', xy=(25, 60), xytext=(30, 70), 
              arrowprops=dict(facecolor='black', shrink=0.05)) 
ax2.bar(x2,y2)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()

```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )

fig.tight_layout()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4), layout='constrained')
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )

fig.tight_layout()
```

```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")

```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")



```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.errorbar(x, y2, color = "red",fmt='none', yerr=y2_sd)
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.errorbar(x, y2, color = "red",fmt='-', yerr=y2_sd)
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "blue")
ax.errorbar(x, y2, color = "yellow",fmt='-', yerr=y2_sd)
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "blue")
ax.errorbar(x, y2, color = "yellow",fmt='-', yerr=y2_sd,
            ecolor='green', capsize=6)
```

```python
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "blue")
ax.errorbar(x, y2, color = "yellow",fmt='-', yerr=y2_sd,
            ecolor='green', capsize=6, capthick=5)
```
