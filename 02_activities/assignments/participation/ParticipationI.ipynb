```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')

# use the function scatterplot() to make a scatterplot
sns.scatterplot(x=df["sepal_length"], y=df["sepal_width"])





sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species",    
    style="species"   
) 
# Overall allows better visual distinction between species and grouping


plt.title("Sepal Length vs Width by Iris Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.tight_layout()
```
